```python
df_test = pd.read_csv('test.csv')
df_test = pd.get_dummies(df_test, columns=['taxi_type', 'customer_score_confidence', 'drop_location_type', 'sex'], dummy_na=True)
```

```python
import math
from sklearn import preprocessing
# Preprocessing fxns
def pp0(df_test):
    return df_test

def pp1(df_test):
    df_test = df_test.fillna(df_test.mean())
    return df_test

def pp2(df_test):
    df_test['anon_var_1'] = df_test['anon_var_1'].apply(lambda x: 0 if math.isnan(x) else x)    
    #df_test['customer_score'] = df_test['customer_score'].apply(lambda x: -1 if math.isnan(x) else x)    
    
    df_test = df_test.fillna(df_test.mean())
    return df_test

def pp3(df_test):
    df_test['anon_var_1'] = df_test['anon_var_1'].apply(lambda x: 0 if math.isnan(x) else x)    
            
    df_test = df_test.fillna(df_test.mean())
    
    train_norm = df_test[['distance', 'customer_score', 'ratings_given_by_cust', 'months_of_activity'
                          , 'num_of_cancelled_trips']]
    #print(norm.head())
    std_scale = preprocessing.MinMaxScaler().fit(train_norm)
    x_train_norm = std_scale.transform(train_norm)
    
    training_norm_col = pd.DataFrame(x_train_norm, index=train_norm.index, columns=train_norm.columns) 
    df_test.update(training_norm_col)
    
    #print (df_test.head())
    
    return df_test

def pp4(df_test):
    
    df_test = df_test.fillna(df_test.median())

    vector = []
    
    for idx, line in df_test.iterrows():
        if math.isnan(line['customer_score']):
            vector.append(-1)
        elif line['customer_score_confidence_A'] == 1:
            vector.append(line['customer_score']+10)
        elif line['customer_score_confidence_B'] == 1:
            vector.append(line['customer_score']+5)
        else:
            vector.append(line['customer_score'])
    
    
    df_test = df_test.fillna(df_test.mean())
    
    train_norm = df_test[['distance', 'customer_score', 'ratings_given_by_cust', 'months_of_activity'
                          , 'num_of_cancelled_trips']]
    #print(norm.head())
    std_scale = preprocessing.MinMaxScaler().fit(train_norm)
    x_train_norm = std_scale.transform(train_norm)
    
    training_norm_col = pd.DataFrame(x_train_norm, index=train_norm.index, columns=train_norm.columns) 
    df_test.update(training_norm_col)
    
    #print(norm.head())
    
    training_norm_col = pd.DataFrame(vector, columns=['customer_score']) 
    df_test.update(training_norm_col)
    
    #print (df_test.head())
    
    return df_test
```

```python
import xgboost as xgb
xgb_model = xgb.XGBClassifier(objective="multi:softprob", n_jobs=6
                              , booster='gbtree', learning_rate=0.1)
```

```python
from sklearn.model_selection import RandomizedSearchCV

def report_best_scores(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
from scipy.stats import uniform, randint
params = {
    "colsample_bytree": uniform(0.7, 0.3),
    "gamma": uniform(0, 0.5),
    "learning_rate": uniform(0.03, 0.3), # default 0.1 
    "max_depth": randint(2, 6), # default 3
    "n_estimators": randint(100, 150), # default 100
    "subsample": uniform(0.6, 0.4)
}
search = RandomizedSearchCV(xgb_model, param_distributions=params, cv=3,  n_jobs=1, return_train_score=True)
search.fit(trainX, trainY)
```

```python
report_best_scores(search.cv_results_, 1)
```
